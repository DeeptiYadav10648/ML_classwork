```python
#dimensionality reduction
from sklearn.decomposition import PCA
model = PCA(n_components=5)
```

```python
#split x and y
#store 536 input labels in x and output label in y
y = data['Label']
data.drop('Label', axis = 1,inplace=True)
x = data
```

```python
y.unique()
```

```python
x_pca = model.fit_transform(x)
```

```python
print(model.explained_variance_ratio_)
```

```python
#sum of explained_variance_ratio
print(sum(model.explained_variance_ratio_))
```

```python
#train test splitting
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x_pca,y,test_size=0.3)
```

```python
#print x_train and y_train
#print x_train.head
x_Train = pd.DataFrame(x_train,columns=['PC1','PC2','PC3','PC4','PC5'])
x_Train.head()
```

```python
y_train.head()
```

```python
#import logistic regression
from sklearn.linear_model import LogisticRegression
model1 = LogisticRegression()
```

```python
model1.fit(x_train,y_train)
```

```python
#evalution metrics accuracy , z score
from sklearn.metrics import accuracy_score,f1_score
y_pred = model1.predict(x_test)
print(accuracy_score(y_test,y_pred))
print(f1_score(y_test,y_pred))
```

```python
#visualization before applying clustering
import matplotlib.pyplot as plt
plt.scatter(x_Train['PC1'],x_Train['PC2'],c=y_train)
plt.show()
```

```python
#Clustering
from sklearn.cluster import KMeans
model2 = KMeans()   #for no. of clusters we use n_clusters ,,
```

```python
#use x_Train for clustering
kmeans = KMeans(n_clusters=2, init='k-means++',copy_x=False,algorithm='lloyd').fit(x_Train)
```

```python
#silhoutte score
from sklearn.metrics import silhouette_score
silhouette_score(x_Train,kmeans.labels_)
```

```python
#visualize kmeans clustering
import matplotlib.pyplot as plt
plt.scatter(x_Train['PC1'],x_Train['PC2'],c=kmeans.labels_)
plt.show()
```

```python
#DBScan
from sklearn.cluster import DBSCAN
model3 = DBSCAN()
```

```python
dbscan = DBSCAN(eps=0.4,min_samples=5,n_jobs=-1,).fit(x_Train)
```

```python
#evaluation metric in dbscan
from sklearn.metrics import silhouette_score
print(silhouette_score(x_Train,dbscan.labels_))
```
